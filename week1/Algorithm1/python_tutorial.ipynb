```python
# Features of Python's Functions
# 1. Able to assign function to variable
def greet(name):
    return "Hi, {}".format(name)

g = greet
g('Korea')
```

```python
# 2. Able to define function inside another function
def greeting(name):
    def greet_message():
        return 'Hi'
    return "{} {}".format(greet_message(), name)

greeting('Korea')
```

```python
# 3. Python function can be an argument of another function
def change_name_greet(func):
    name = "US"
    return func(name)

change_name_greet(greet)
```

```python
# 4. Python function can be a return value of another function
def uppercase(func): 
    def wrapper(name):
        result = func(name)
        return result.upper()
    return wrapper # 내부 정의된 함수를 밖으로 꺼낸다?

new_greet = uppercase(greet)
new_greet("EU")
```

```python
# '=' vs Shallow Copy vs Deep Copy
## '=' : Not copy, just making binding between objects(same id)
### y -> a variable that refers to the same address of the object x referring to
### Refer to the same object, so share same results of update & delete & addition
x = [1,2,3]
y = x
print(id(x), id(y))
del y[1]
print(x)
y.append(2)
print(x)
```

```python
## Exception : immutable objects
### All integers have their own id 
a = 10
b = a
print(b)    # 10
print(id(a), id(b))
b += 2
print(a, b)
print(id(a), id(b))
b -= 2
print(a, b)
print(id(a), id(b))
b = 2 # refers to 2, not referring to a anymore
print(b)    # abc
print(a)    # 10
print(id(a), id(b), id(2))
```

```python
## Shallow copy(different id)
### A shallow copy constructs a new compound object
### and then (to the extent possible) inserts references into it to the objects found in the original
### but internal objects(mutable) inside the compound object refers to same object
import copy as cp
a = [1, [1,2,3]]
b = cp.copy(a)
print('-------shallow copy-------')
print(b)
del b[0]
print('-------changed copied list-------')
print(b)
print('-------original list(no change)-------')
print(a)
# del b[1]
c = cp.copy(a)
c.append(4)
print('-------the original object still unchanged even for append-------')
print(a)
print('-------add 4 to the second item in the list (inner list)-------')
del c[2]
c[1].append(4)
print(c)
print(a)
### In case of b, element deletion is not reflected because one element(int) is immutable
### In case of c, the inner list is mutable, referring to the same object ('=')
```

```python
## Deep copy(different id)
### A deep copy constructs a new compound object
### and then, recursively, inserts copies into it of the objects found in the original
### Even for inner compund objects(mutable)

a = [0, [1,2,3]]
b = cp.deepcopy(a)
print('-------deep copy-------')
print(b)
b[0] = 100
b[1].append(4)
print('-------original list unchanged-------')
print(b)
print(a)
```

```python
# Packing & Unpacking
numbers = (1, 2, 3, 4, 5) # packing
(a, b, c, d, e) = numbers # unpacking
print(a, b, c, d, e)
```

```python
## In '=' statements, parentheses are often omitted
numbers = 1, 2, 3, 4, 5 # packing
a, b, c, d, e = numbers # unpacking
a, b, c, _, e = numbers # _
a, _, _, d, _ = numbers
print(a, d)
## A variable with * is assigned other remaining elements after unpacking
a, b = numbers[0], numbers[1]
rest = numbers[2:] # rest = (3, 4, 5)
print('-------simple rest-------')
print(a, b, rest)
print('-------assign remaining elements after unpacking to rest-------')
a, b, *rest = numbers
print(a, b, rest)
print('-------print only rest-------')
a, *rest, e = numbers
print(rest)
```

```python
## Unpacking for arguments and keyword arguments
def date_to_str(y, m, d):
    print (str(y) + '년 '+ str(m) + '월 ' + str(d) + '일')

date = (1917, 9, 4)
date_dict = {'y': 2018, 'm': 7, 'd': 28}
date_to_str(date[0], date[1], date[2])
date_to_str(*date) # unpacking arguments
date_to_str(*[2018, 7, 7])
date_to_str(2018, *(7, 18))
date_to_str(**date_dict) # unpacking keyword arguments
```

```python
dict2 = {'y': 2018, 'm': 7, 'd': 28}
date_to_str(**dict2)
```

```python
def draw_point(x, y):
    print(x, y)
    
point_foo = (3, 4) # tuple argument packing
point_bar = {'y': 3, 'x': 2} # dictionary keyword argument packing

draw_point(*point_foo) # tuple unpacking
draw_point(**point_bar) # dictionary unpacking(mapping unpacking)
```

```python
## Input through mapping
def draw_tuple(*args): # input after converting to tuple type
    print(args)
    
def draw_dict(**kwargs): # input after converting to dictionary type
    print(kwargs)
    for key in kwargs:
        print(key, kwargs[key])
```

```python
print("---empty---")
draw_tuple()
print("---one element---")
draw_tuple(1)
print("---str---")
draw_tuple("hello", 'world')
print("---dictionary---")
draw_dict(Korea=80, US=85, EU=90)

d = {'Sun':100, 'Moon':50}
draw_dict(**d)
```

```python
def date_to_str(y, m, d, **kwargs):
    result = str(y) + '년 '+ str(m) + '월 ' + str(d) + '일'
    print(kwargs)
    if 'h' in kwargs:
        result += ' ' + str(kwargs['h']) + '시'
    print(result)
    
date_to_str(2018, 7, 30)
date_to_str(2018, 7, 30, h=2, ms=4) # h=2, ms=4 -> {'h':2, 'ms':4}
```

```python
# Generator
def gen():
## When generator function is called at first,
## it returns the value from the first yield value encountered during execution.
## When called again, statements are executed from the next yield statement to the next yield statement
    yield 1
    yield 2
    yield 3
    
g = gen()
print(type(g))
n = next(g); print(n)  # 1
n = next(g); print(n)  # 2
n = next(g); print(n)  # 3

for x in gen():
    print(x)
```

```python
# Generator Expression

g = (i*i for i in range(1001))
print(type(g))
for itr in range(10):
    val = next(g)
    print(val)

print("------")
for itr in g: # started from the next yield value
    print(itr)
```

```python
def gener(n):
    i=0
    while i < n:
        yield i
        i += 1
        
a = gener(5)
print(type(a))

for x in gener(5): # A generator is also an iterator and yield makes the funcation a generator
    print(x)


# 1. for statement is executed, and generator function is first called
# 2. A generator function is executed in the same procedure as general functions
# 3. During the execution, yield is encountered in while statement.
#    Then, similar with return in functions, it returns the statement that called the function.
#    Here, the first i value, 0, is returned.
#    However, return does not terminate the function, but maintains it.
# 4. In the x value, the 0 value passed through yield is stored and then printed.
#    Then, the function is called again by for statement.
# 5. In this case, the function does not start from the beginning, but starts with the statement after yield.
#    So i + = 1 is executed and i is incremented by 1 and becomes 1.
# 6. Since it is still inside while statement, the i value of 1 is passed by yield statement.
# 7. The x value gets 1 and is printed. Repeat afterwards.
```

```python
# Advantages of generators in python : efficient use of memory and fast processing speed
# Because generators load values into memory whenever values are accessed one after the other through the next () method

import sys
print(sys.getsizeof( [i for i in range(100)] ))
print(sys.getsizeof( [i for i in range(1000)] ))
print(sys.getsizeof( (i for i in range(100)) ) )
print(sys.getsizeof( (i for i in range(1000)) ))
```

```python
from itertools import islice
from itertools import cycle
colors = cycle(['red', 'white', 'blue'])
print(next(colors))
print(next(colors))
print(next(colors))
print(next(colors))
print('-------from infinite sequence to finite sequence-------')
limited = islice(colors, 0, 4)
for x in limited:
    print(x)
```

```python
def fibonacci():
    prev, curr = 0, 1
    while True:
        yield curr
        prev, curr = curr, prev + curr
        
f = fibonacci()
list(islice(f, 0, 10))
```

```python
def fibo(n):
    if n < 2:
        return n
    return fibo(n-1) + fibo(n-2)
        
[fibo(i) for i in range(11)]
```

```python
memo = {1:1, 2:1}
def fibo2(n):
    if n == 0:
        return 0
    if n not in memo:
        for itr in range(3,n+1):
            memo[itr] = fibo2(itr-2) + fibo2(itr-1)
    return memo[n]

[fibo2(i) for i in range(11)]
```
